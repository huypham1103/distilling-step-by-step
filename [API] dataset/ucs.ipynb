```python
class UCS:
    def __init__(self) -> None:
        pass
    
    def token_probabilities(self, generations):
        """
        Compute the token probabilities for each word in the generations.
        """
        total_tokens = sum(len(gen.split()) for gen in generations)
        token_freq = {}
        for gen in generations:
            for token in gen.split():
                token_freq[token] = token_freq.get(token, 0) + 1
        return {token: freq / total_tokens for token, freq in token_freq.items()}
    
    def compute_UCS(self, generation, token_probs):
        """
    Compute the unigram consistency score (UCS) for a generation.
    """
        return sum(token_probs.get(token, 0) for token in generation.split())
    
    def compute_WUCS(self, generation, token_probs):
        """
        Compute the weighted unigram consistency score (WUCS) for a generation.
        """
        return sum(token_probs.get(token, 0) ** 2 for token in generation.split())
    
    def compute_Consensus_WUCS(self, generation, token_probs):
        """
        Compute the Consensus-WUCS for a generation.
        """
        WUCS = self.compute_WUCS(generation, token_probs)
        mean_log_prob = sum(
            math.log(token_probs.get(token, 1)) for token in generation.split()
        ) / len(generation.split())
        return WUCS * math.exp(mean_log_prob)
```

```python
ucs = UCS()
```

```python
def get_highest_rationales(lines, type='UCS'):
    # Example usage
    res = []
    for line in lines:
        line_res = {}
        generations = line
        token_probs = ucs.token_probabilities(generations)

        for j in range(len(line)):
            generation_to_evaluate = line[j]
            if type=='UCS':
                consensus_wucs_score = ucs.compute_UCS(generation_to_evaluate, token_probs)
            elif type=='WUCS':
                consensus_wucs_score = ucs.compute_WUCS(generation_to_evaluate, token_probs)
            elif type=='Consensus-WUCS':
                consensus_wucs_score = ucs.compute_Consensus_WUCS(generation_to_evaluate, token_probs)
            
            line_res[line[j]] = consensus_wucs_score
        
        # get the rationale with the highest UCS score
        max_key = max(line_res, key=line_res.get)
        res.append(max_key)

    return res
```

```python
def _cal_token_probabilities(generations):
    """
    Compute the token probabilities for each word in the generations.
    """
    total_tokens = sum(len(gen.split()) for gen in generations)
    token_freq = {}
    for gen in generations:
        for token in gen.split():
            token_freq[token] = token_freq.get(token, 0) + 1
    return {token: freq / total_tokens for token, freq in token_freq.items()}
```

```python
def calculate_individual_ucs_scores(sentences):
    """
    Calculate the Unigram Consistency Score (UCS) for each sentence in a list,
    compared to the rest of the sentences.

    Args:
    sentences (list of str): The list of sentences to calculate the UCS scores for.

    Returns:
    list of float: The calculated UCS scores for each sentence.
    """
    unigram_counts = [Counter(sentence.split()) for sentence in sentences]
    scores = []
    for i, count_a in enumerate(unigram_counts):
        total_overlap = 0
        num_pairs = 0
        for j, count_b in enumerate(unigram_counts):
            if i != j:
                overlap = sum((count_a & count_b).values())
                total_overlap += overlap
                num_pairs += 1
        score = total_overlap / num_pairs if num_pairs > 0 else 0
        scores.append(score)
    return scores

# Calculate UCS score for each sentence
individual_ucs_scores = calculate_individual_ucs_scores(generations)
individual_ucs_scores
```

```python
def calculate_wucs_scores(sentences, token_probabilities):
    """
    Calculate the Weighted Unigram Consistency Score (WUCS) for each sentence.

    :param sentences: A list of sentences.
    :param token_probabilities: A dictionary mapping tokens to their probabilities.
    :return: A list of WUCS scores for each sentence.
    """
    tokenized_sentences = [tokenize(sentence) for sentence in sentences]
    unigram_counts = [Counter(tokens) for tokens in tokenized_sentences]
    total_unigrams = sum((Counter(tokens) for tokens in tokenized_sentences), Counter())

    wucs_scores = []
    for i, gen in enumerate(tokenized_sentences):
        weighted_overlap = 0
        for j, gen_other in enumerate(tokenized_sentences):
            if i != j:
                for token in set(gen).intersection(set(gen_other)):
                    overlap = min(unigram_counts[i][token], unigram_counts[j][token])
                    weighted_overlap += overlap * token_probabilities.get(token, 0) / (total_unigrams[token] if total_unigrams[token] > 0 else 1)
        wucs_scores.append(weighted_overlap)

    return wucs_scores

# Example usage
sentences = ["The sky is blue.", "The sky looks blue.", "Blue is the color of the sky."]
token_probabilities = _cal_token_probabilities(sentences)
wucs_scores = calculate_wucs_scores(sentences, token_probabilities)
wucs_scores

```

```python
def tokenize(sentence):
    """Simple tokenizer to split a sentence into words."""
    return sentence.split()

def calculate_cwucs_scores(sentences, token_probabilities):
    """
    Calculate the Cumulative Weighted Unigram Consistency Score (CWUCS) for each sentence.

    :param sentences: A list of sentences.
    :param token_probabilities: A dictionary mapping tokens to their probabilities.
    :return: A list of CWUCS scores for each sentence.
    """
    tokenized_sentences = [tokenize(sentence) for sentence in sentences]
    unigram_counts = [Counter(tokens) for tokens in tokenized_sentences]
    total_unigrams = sum((Counter(tokens) for tokens in tokenized_sentences), Counter())

    cwucs_scores = []
    for i, gen in enumerate(tokenized_sentences):
        cumulative_weighted_overlap = 0
        for j, gen_other in enumerate(tokenized_sentences):
            if i != j:
                for token in set(gen).intersection(set(gen_other)):
                    weighted_overlap = min(unigram_counts[i][token], unigram_counts[j][token]) * token_probabilities.get(token, 0) / (total_unigrams[token] if total_unigrams[token] > 0 else 1)
                    cumulative_weighted_overlap += weighted_overlap
        mean_log_prob = np.mean([np.log(token_probabilities.get(token, 1)) for token in gen])
        cwucs_scores.append(cumulative_weighted_overlap * np.exp(mean_log_prob))

    return cwucs_scores

# Example usage
sentences = ["The sky is blue.", "The sky looks blue.", "Blue is the color of the sky."]
token_probabilities = _cal_token_probabilities(sentences)
cwucs_scores = calculate_cwucs_scores(sentences, token_probabilities)
cwucs_scores

```
