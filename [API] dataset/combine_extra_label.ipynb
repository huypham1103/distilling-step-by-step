```python
import pandas as pd
```

```python
NAMES = ['if_else', 'contrastive', 'neutral', 
         'consensus', 'causal', 'comparative', 
         'historical', 'ucs_score', 'wucs_score', 
         'consensus_wucs_score', 'concat_all', 'best_prompt']
```

```python
for i in NAMES:
    for j in NAMES:
        csv_1 = pd.read_csv(f'{i} - full.csv')
        csv_1.set_index('question', inplace=True)

        csv_2 = pd.read_csv(f'{j} - full.csv')
        csv_2.set_index('question', inplace=True)
        
        csv_1['rationales_2'] = csv_2['rationales']
        csv_1['LLM_answer_2'] = csv_2['LLM_answer']
        
        # if LLM_answer different from LLM_answer_2, then fill rationales_2 with rationales
        csv_1.loc[csv_1['LLM_answer'] != csv_1['LLM_answer_2'], 'rationales_2'] = csv_1['rationales']

        csv_1.reset_index(inplace=True)
        # Export to csv
        csv_1.to_csv(f'extra_rationales/{i}_{j}.csv', index=False)
```
